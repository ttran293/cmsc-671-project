```python
import pandas as pd

# URL of the CSV file
df_url = "https://media.githubusercontent.com/media/ttran293/cmsc-671-project/refs/heads/main/clean_dataset/gold_output.csv"

df = pd.read_csv(df_url, encoding="utf-8")
```

```python
records = []
for _, row in df.iterrows():
    original_text = str(row["original_text"])
    summarized_text = str(row["summarized_text"])

    instruction = "Summarize the following text."

    if not original_text or not summarized_text:
        continue

    records.append({
        "instruction": instruction,
        "input": original_text,
        "output": summarized_text
    })

with open("train.jsonl", "w", encoding="utf-8") as f:
    for r in records:
        f.write(json.dumps(r, ensure_ascii=False) + "\n")
```
